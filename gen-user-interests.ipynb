```python
import pandas as pd

grid_size = 60

records = []
for max_lat in range(90, -90, -grid_size):
    min_lat = max_lat - grid_size
    for min_lon in range(-180, 180, grid_size):
        max_lon = min_lon + grid_size
        records.append([max_lat, min_lat, max_lon, min_lon])

df_grid = pd.DataFrame(records, columns=["max_lat", "min_lat", "max_lon", "min_lon"])
df_grid
```

```python
records = []
for end_year in range(2020, 2000, -10):
    start_year = end_year - 9
    records.append([start_year, end_year])
df_year = pd.DataFrame(records, columns=['start_year', 'end_year'])
df_year
```

```python
df = pd.merge(df_grid, df_year, how='cross')
df
```

```python
def gen_resolution(row):
    min_lat = row["min_lat"]
    start_year = row["start_year"]
    if start_year >= 2011:
        if min_lat >= 0:
            return "day"
        else:
            return "month"
    else:
        if min_lat >= 0:
            return "month"
        else:
            return "year"

df["time_resolution"] = df.apply(gen_resolution, axis=1)
df
```

```python
df.to_csv("user_interests.csv", index=False)
```
